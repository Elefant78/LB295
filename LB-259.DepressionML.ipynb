## Aufgabe 4.3

Nun muss ich noch die andere Art von Evaluation machen, nämlich die Wahrheitsmatrix.

Um eine Wahrheitsmatrix (confusion matrix) zu erstellen sowie Sensitivität(recall) und Spezifizität(prescision) zu berechnen, brauche ich logischerweise ein Modell, mit dem ich dies überhaupt evaluieren kann. Da dieses Modell aktuell Regression (z. B. Vorhersage des CGPA) verwendet, müssen wir zuerst eine sinnvolle Bedingung definieren, um daraus eine binäre Klassifikation zu machen.

```python
# Binäre Zielvariable erstellen
y_test_binary = (y_test >= 7.5).astype(int)
y_pred_binary = (y_pred >= 7.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test_binary, y_pred_binary)
print("Konfusionsmatrix:")
print(cm)
```

```python
TP = cm[1, 1]
FN = cm[1, 0]
FP = cm[0, 1]
TN = cm[0, 0]

# Sensitivität (Recall): Wie viele der echten Positiven wurden erkannt?
recall = TP / (TP + FN)

# Spezifizität: Wie viele der echten Negativen wurden korrekt erkannt?
specificity = TN / (TN + FP)

print(f"Sensitivität: {recall:.2f}")
print(f"Spezifizität: {specificity:.2f}")
```
