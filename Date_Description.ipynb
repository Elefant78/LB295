```python
import pandas as pd

DS = pd.read_csv("Student Depression Dataset.csv") 
DS.head(20) 
```

```python
median_FS = DS["Financial Stress"].median()
print("Der Median von Financial Stress liegt bei:", median_FS)
```

```python
mean_CGPA=DS["CGPA"].mean()
print("Der Durschnitt von CGPA liegt bei: ", mean_CGPA)
```

```python
median_CGPA=DS["CGPA"].median()
print("Der Median von CGPA liegt bei:", median_CGPA)
```

```python
#mit diesem Befehl kann ich  den Mittelwert, Median, Standardabweichung, Min, Max aller nummerischen Daten darstellen.(Median ist 50%)
print(DS.describe())
```

```python
import matplotlib.pyplot as plt


plt.hist(DS["Sleep Duration"], bins=10, color='skyblue', edgecolor='black')

plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Hufigkeit")
plt.title("Histogramm der Schlafdauer")
plt.show()
```
